```python
movies = pd.read_csv("ml-latest-small/movies.csv")
ratings = pd.read_csv("ml-latest-small/ratings.csv")

print("movies loaded", movies.shape)
print("ratings loaded",ratings.shape)
```

```python
def search_movies(keyword):
    results = movies[movies["title"].str.contains(keyword, case=False, na=False)]
    return results[["movieId", "title", "genres"]].head(20)
```

```python
search_movies("Spider")
```

```python
my_ratings = {
    "Matrix, The (1999)": 5,
    "Dark Knight, The (2008)": 5,
    "Iron Man (2008)": 4,
    "Avengers, The (2012)": 4,
    "Toy Story (1995)": 2
}

my_ratings
```

```python
rated_movies = movies[movies["title"].isin(my_ratings.keys())]

rated_movies
```

```python
rated_movies = rated_movies.copy()

rated_movies["my_rating"] = rated_movies["title"].map(my_ratings)

rated_movies
```

```python
movies_clean = movies.copy()

movies_clean["genres_clean"] = movies_clean["genres"].str.replace("|", " ", regex=False)

movies_clean[["title", "genres", "genres_clean"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

print("Text vectorizer imported")
```

```python
vectorizer = TfidfVectorizer()

genre_matrix = vectorizer.fit_transform(movies_clean["genres_clean"])

print("Genre matrix created successfully")
print("Shape:", genre_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

print("Similarity tool imported")
```

```python
liked_profile = np.zeros(genre_matrix.shape[1])

for movie_index, rating in zip(rated_movies.index, rated_movies["my_rating"]):
    liked_profile += genre_matrix[movie_index].toarray()[0] * rating

print("Taste profile created")
print("Profile shape:", liked_profile.shape)
```

```python
similarity_scores = cosine_similarity([liked_profile], genre_matrix).flatten()

movies_clean["recommendation_score"] = similarity_scores

movies_clean[["title", "genres", "recommendation_score"]].sort_values(
    by="recommendation_score",
    ascending=False
).head(10)
```

```python
already_rated_titles = list(my_ratings.keys())

recommendations = movies_clean[
    ~movies_clean["title"].isin(already_rated_titles)
].sort_values(
    by="recommendation_score",
    ascending=False
)

recommendations[["title", "genres", "recommendation_score"]].head(20)
```

```python
def recommend_movies(my_ratings, top_n=20):
    rated_movies = movies_clean[movies_clean["title"].isin(my_ratings.keys())].copy()

    if rated_movies.empty:
        return "No matching movies found. Use search_movies() to check exact titles."

    rated_movies["my_rating"] = rated_movies["title"].map(my_ratings)

    liked_profile = np.zeros(genre_matrix.shape[1])

    for movie_index, rating in zip(rated_movies.index, rated_movies["my_rating"]):
        liked_profile += genre_matrix[movie_index].toarray()[0] * rating

    similarity_scores = cosine_similarity([liked_profile], genre_matrix).flatten()

    results = movies_clean.copy()
    results["recommendation_score"] = similarity_scores

    already_rated_titles = list(my_ratings.keys())

    recommendations = results[
        ~results["title"].isin(already_rated_titles)
    ].sort_values(
        by="recommendation_score",
        ascending=False
    )

    return recommendations[["title", "genres", "recommendation_score"]].head(top_n)
```

```python
recommend_movies(my_ratings, top_n=55)
```

```python
movie_stats = ratings.groupby("movieId").agg(
    average_rating=("rating", "mean"),
    rating_count=("rating", "count")
).reset_index()

movie_stats.head()
```

```python
movies_clean = movies_clean.merge(movie_stats, on="movieId", how="left")

movies_clean[["movieId", "title", "genres", "average_rating", "rating_count"]].head()
```

```python
movies_clean["average_rating"] = movies_clean["average_rating"].fillna(0)
movies_clean["rating_count"] = movies_clean["rating_count"].fillna(0)

movies_clean[["title", "average_rating", "rating_count"]].head()
```

```python
movies_clean["public_rating_score"] = movies_clean["average_rating"] / 5

movies_clean["popularity_score"] = np.log1p(movies_clean["rating_count"])
movies_clean["popularity_score"] = movies_clean["popularity_score"] / movies_clean["popularity_score"].max()

movies_clean[[
    "title",
    "average_rating",
    "rating_count",
    "public_rating_score",
    "popularity_score"
]].head()
```

```python
movies_clean["final_score"] = (
    movies_clean["recommendation_score"] * 0.70 +
    movies_clean["public_rating_score"] * 0.20 +
    movies_clean["popularity_score"] * 0.10
)

movies_clean[[
    "title",
    "genres",
    "recommendation_score",
    "average_rating",
    "rating_count",
    "final_score"
]].sort_values(
    by="final_score",
    ascending=False
).head(20)
```

```python
already_rated_titles = list(my_ratings.keys())

final_recommendations = movies_clean[
    ~movies_clean["title"].isin(already_rated_titles)
].sort_values(
    by="final_score",
    ascending=False
)

final_recommendations[[
    "title",
    "genres",
    "recommendation_score",
    "average_rating",
    "rating_count",
    "final_score"
]].head(20)
```

```python
def recommend_movies_v2(my_ratings, top_n=20):
    rated_movies = movies_clean[movies_clean["title"].isin(my_ratings.keys())].copy()

    if rated_movies.empty:
        return "No matching movies found. Use search_movies() to check exact titles."

    rated_movies["my_rating"] = rated_movies["title"].map(my_ratings)

    liked_profile = np.zeros(genre_matrix.shape[1])

    for movie_index, rating in zip(rated_movies.index, rated_movies["my_rating"]):
        liked_profile += genre_matrix[movie_index].toarray()[0] * rating

    similarity_scores = cosine_similarity([liked_profile], genre_matrix).flatten()

    results = movies_clean.copy()
    results["recommendation_score"] = similarity_scores

    results["public_rating_score"] = results["average_rating"] / 5

    results["popularity_score"] = np.log1p(results["rating_count"])
    results["popularity_score"] = results["popularity_score"] / results["popularity_score"].max()

    results["final_score"] = (
        results["recommendation_score"] * 0.70 +
        results["public_rating_score"] * 0.20 +
        results["popularity_score"] * 0.10
    )

    already_rated_titles = list(my_ratings.keys())

    recommendations = results[
        ~results["title"].isin(already_rated_titles)
    ].sort_values(
        by="final_score",
        ascending=False
    )

    return recommendations[[
        "title",
        "genres",
        "recommendation_score",
        "average_rating",
        "rating_count",
        "final_score"
    ]].head(top_n)
```

```python
recommend_movies_v2(my_ratings, top_n=20)
```

```python
clean_recommendations = recommend_movies_v2(my_ratings, top_n=20).copy()

clean_recommendations["recommendation_score"] = clean_recommendations["recommendation_score"].round(3)
clean_recommendations["average_rating"] = clean_recommendations["average_rating"].round(2)
clean_recommendations["final_score"] = clean_recommendations["final_score"].round(3)

clean_recommendations.reset_index(drop=True)
```

```python
clean_recommendations.to_csv("my_movie_recommendations.csv", index=False)

print("Recommendations saved successfully")
```
